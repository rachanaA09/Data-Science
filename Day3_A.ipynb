### Bar Plot

```python
enrollment=[104,112,256,178,126,180]
courses=['C' , 'Java', 'Python', 'R', 'PHP', 'javaScript']
```

```python
plt.xlabel('courses')
plt.ylabel('Enrolied Students')
plt.grid()
plt.bar(courses, enrollment, color='brown')
```

```python
plt.xlabel('courses')
plt.ylabel('Enrolied Students')
plt.grid()
plt.barh(courses, enrollment, color='brown')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
import os
os.getcwd()
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
```

```python
x=df['YearsExperience'].values
y=df['Salary']
```

```python
x
```

```python
y;
```

```python
x=x.reshape(30, 1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

#### create the object of linear regression 

```python
regressor = LinearRegression()
```

#### train the algorithm

```python
regressor.fit(x, y)
```

```python
regressor.predict([[20]])
```

#### predict the unseen data 

```python
regressor.predict([[15]])
```

```python
regressor.score(x, y)
```
